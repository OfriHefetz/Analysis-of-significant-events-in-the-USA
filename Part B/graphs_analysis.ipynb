```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import os

start_path = 'C:/Users/gony/Desktop/'
```

## Analyzing 9/11 graphs

```python
count_nodes_911= []
count_edges_911 = []
weights_9_11 = []
path = start_path+'data/new/9_11_2001/'

for f in os.listdir(path):
    df = pd.read_csv(path + f)
    list_nodes = df.source.values + df.target.values
    count_nodes_911.append(len(set(list_nodes)))
    count_edges_911.append(len(df))
    weights_9_11.extend(df.weight.values)
```

```python
plt.hist(count_nodes_911)
plt.title('histogram of nodes in 9/11 graphs')
plt.show()

plt.hist(count_edges_911)
plt.title('histogram of edges in 9/11 graphs')
plt.show()
```
