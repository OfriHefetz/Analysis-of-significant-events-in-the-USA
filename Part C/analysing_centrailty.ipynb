```python
import os

import pandas as pd
import glob
import matplotlib.pyplot as plt
import networkx as nx
import random
start = 'C:/Users/gony/Desktop/data/'
```

 since we are interested in identifying nodes that are frequently mentioned or highly connected, Degree Centrality is suitable

```python
def plot_top_n(top_dict, colors, title=None):
    # Extract the words and their centrality scores
    words = list(top_dict.keys())
    scores = list(top_dict.values())
    # Create a list of colors
    # colors = ['#8DD3C7', '#FFFFB3', '#BEBADA', '#FB8072', '#80B1D3', '#FDB462', '#B3DE69', '#FCCDE5', '#D9D9D9', '#BC80BD',
    #       '#CCEBC5', '#FFED6F', '#1F78B4', '#FFBC79', '#A6CEE3', '#AFB0D2', '#FFBB78', '#2D2D2D', '#C49C94', '#FB9A99']
    # random.shuffle(colors)  # Shuffle the list of colors randomly

    # Create the line plot
    plt.figure(figsize=(10, 6))  # Set the figure size

    # Plot the lines for each word with different colors
    for i in range(len(words)):
        plt.plot(scores[i], label=words[i], linewidth=2, color=colors[i % len(colors)])  # Use modulo operator to cycle through colors
    output = [i for i in range(-5,16)]
    plt.xlabel('Time')  # X-axis label
    plt.ylabel('Centrality Score')  # Y-axis label
    plt.title('Centrality Scores Over Time')  # Title of the chart
    plt.xticks(range(len(output)), output)
    plt.axvline(x=5, color='black', linestyle='dashed')
    plt.legend(loc='best')
    plt.title(title)
    plt.tight_layout()
    plt.show()  # Display the chart
```

# 9/11 terror attack - September 2001
#### Finding the top 20 nodes the day after the event, and examining their centrality through out time

```python
path = start + 'nodes_centrality_9_11_2001/'

day = pd.read_csv(path + 'centrality_11.09.2001.csv')
top_20_node= day.nlargest(n=20, columns=['degree_cent'])['node'].tolist()
top_20_dict = {}
for f in os.listdir(path):
    if 'community' not in f:
        day = pd.read_csv(path+f)
        for j in top_20_node:
            if j in day['node'].values:
                value = float(day[day['node']==j]['degree_cent'])
            else:
                value = 0
            if j in top_20_dict.keys():
                top_20_dict[j].append(value)
            else:
                top_20_dict[j] = [value]

colors = ['#00204A', '#072E5A', '#0D3E6B','#135D7C', '#1A6D8C', '#207D9D', '#268DAC', '#2C9DBC', '#33ADEB', '#990000',
          '#33ADEB', '#46B2F2', '#5AB7F6' ,'#B30000', '#6EBDFA', '#CC0000' ,'#E50000', '#FF0000', '#D2D8FF','#E6DDFF']
plot_top_n(top_20_dict, colors, title='9/11')
```
